### 2.3.2 Voter Turnout by Age Group

```python
from pyspark.sql.functions import expr, year, to_date, col, when
import matplotlib.pyplot as plt

df1 = spark.table("databricks_demo.final_project.wa").filter(col("Voters_BirthDate").isNotNull())

# Unpivot General_<year> columns into (election, voted)
years = [2000, 2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018]
stack_expr = "stack({0}, {1}) as (election, voted)".format(
    len(years),
    ", ".join(f"'General_{yr}', General_{yr}" for yr in years)
)
unpivoted = df1.select(expr(stack_expr), "Voters_BirthDate")

# Compute birth_year, election year, age, voted_flag, and assign age_group
processed = (
    unpivoted
      .withColumn("birth_year", year(to_date(col("Voters_BirthDate"), "MM/dd/yyyy")))
      .withColumn("year", expr("CAST(SUBSTR(election, 9) AS INT)"))
      .withColumn("age", col("year") - col("birth_year"))
      .withColumn("voted_flag", when(col("voted") == "Y", 1).otherwise(0))
      .withColumn(
          "age_group",
          when((col("age") >= 20) & (col("age") < 30), "20s")
          .when((col("age") >= 30) & (col("age") < 40), "30s")
          .when((col("age") >= 40) & (col("age") < 50), "40s")
          .when((col("age") >= 50) & (col("age") < 60), "50s")
          .when((col("age") >= 60) & (col("age") < 70), "60s")
          .when((col("age") >= 70) & (col("age") < 80), "70s")
          .otherwise("80+")
      )
)

# Aggregate turnout_rate by (year, age_group)
turnout_by_age_year = (
    processed
      .groupBy("year", "age_group")
      .agg((expr("SUM(voted_flag)") / expr("COUNT(1)")).alias("turnout_rate"))
      .orderBy("year", "age_group")
)

# Convert to pandas-on-Spark and pivot so each age_group is a column
turnout_ps = turnout_by_age_year.to_pandas_on_spark()
pivot = turnout_ps.pivot_table(index=["year"], columns="age_group", values="turnout_rate").sort_index()

# Convert to pandas for plotting
pivot_pd = pivot.to_pandas()

# Plot line trace for each age group across all years
plt.figure(figsize=(8, 5))
for age_grp in sorted(pivot_pd.columns):
    plt.plot(pivot_pd.index, pivot_pd[age_grp], marker='o', label=age_grp)
plt.xlabel("Election Year")
plt.ylabel("Turnout Rate")
plt.title("WA Voter Turnout by Age Group (2000–2018)")
plt.xticks(pivot_pd.index.astype(int))  # Set x-ticks to exact year values
plt.legend(title="Age Group")
plt.grid(True)
plt.tight_layout()
plt.show()
```

The graph demonstrates a line trace of the different age groups for the 2000 to 2018 election years. The 80+ age group demonstrates an upward trend after 2014, narrowing the gap with younger age groups. However, the age groups between 20 to 50 years old demonstrate higher voter turnout for the presidential elections and lower turnout for the midterm elections. The age groups between 60 and 70 years old demonstrate the same pattern, but do not have a significant decrease for the midterm elections. Overall, the 80+ age group is consistent with voting through the midterm and presidential elections.

### Do voters in high-turnout counties also tend to live in higher income areas?

```python
from pyspark.sql.functions import when, col, expr, avg, sum as spark_sum, count as spark_count
from pyspark.ml.feature import StringIndexer, VectorAssembler, StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.ml import Pipeline


years = [2000,2002,2004,2006,2008,2010,2012,2014,2016,2018]


full = spark.table("databricks_demo.final_project.wa")


income_indexer = StringIndexer(
    inputCol="CommercialData_EstimatedHHIncome",
    outputCol="income_idx",
    handleInvalid="keep"
).fit(full)
full_idx = income_indexer.transform(full)


for yr in years:
    full_idx = full_idx.withColumn(
        f"voted_{yr}",
        when(col(f"General_{yr}") == "Y", 1).otherwise(0)
    )


sum_expr = " + ".join(f"SUM(voted_{yr})" for yr in years)
county_feats = (
    full_idx
      .groupBy("County")
      .agg(
        expr(f"({sum_expr}) / (COUNT(*) * {len(years)})")
          .alias("turnout_rate_all"),
        avg("income_idx").alias("avg_income_idx")
      )
      .na.drop(subset=["avg_income_idx"])
)


assembler = VectorAssembler(
    inputCols=["turnout_rate_all", "avg_income_idx"],
    outputCol="raw_features"
)
scaler = StandardScaler(
    inputCol="raw_features",
    outputCol="features",
    withMean=True,
    withStd=True
)
kmeans = KMeans(k=3, seed=42, featuresCol="features")


pipeline = Pipeline(stages=[assembler, scaler, kmeans])


model = pipeline.fit(county_feats)
clusters = model.transform(county_feats)
```

### K-Means Clustering Output Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

clusters_pd = clusters \
    .select("County", "turnout_rate_all", "avg_income_idx", "prediction") \
    .toPandas()


centers = clusters_pd.groupby("prediction")[["avg_income_idx", "turnout_rate_all"]] \
                     .mean() \
                     .reset_index()


fig, ax = plt.subplots(figsize=(8, 6))
scatter = ax.scatter(
    clusters_pd["avg_income_idx"],
    clusters_pd["turnout_rate_all"],
    c=clusters_pd["prediction"],
    cmap="tab10",
    alpha=0.7,
    edgecolor="k"
)
ax.legend(*scatter.legend_elements(), title="Cluster")
for _, row in centers.iterrows():
    ax.scatter(
        row["avg_income_idx"],
        row["turnout_rate_all"],
        marker="X",
        s=200,
        c="black"
    )
    ax.text(
        row["avg_income_idx"],
        row["turnout_rate_all"] + 0.005,
        f"Centroid {int(row['prediction'])}",
        ha="center",
        va="bottom",
        fontweight="bold"
    )

ax.set_xlabel("Average Income Index")
ax.set_ylabel("Turnout Rate (2000–2018)")
ax.set_title("County Clusters: Income vs. Turnout")
ax.grid(True)
plt.tight_layout()
plt.show()


summary = clusters_pd.groupby("prediction")["turnout_rate_all"].mean().reset_index()

fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(
    summary["prediction"],
    summary["turnout_rate_all"],
    color="skyblue",
    edgecolor="k"
)
ax.set_xticks(summary["prediction"])
ax.set_xlabel("Cluster")
ax.set_ylabel("Mean Turnout Rate")
ax.set_title("Mean Turnout Rate by Cluster")
for bar in bars:
    h = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        h + 0.005,
        f"{h:.3f}",
        ha="center",
        va="bottom"
    )
plt.tight_layout()
plt.show()
```

#### Interpretation

We began by transforming every voter’s history into two summary features at the county level: a turnout rate measuring the fraction of all possible general‐election votes cast from 2000 through 2018, and an income index derived by string-indexing the household-income category for each voter and then averaging those indices across the county. Because counties vary in both dimensions, standardizing each feature and feeding them into a three-cluster K-Means model let us see natural groupings that reflect joint patterns of electoral participation and relative wealth.

The clustering produced three distinct county groups. The first cluster, comprising eleven counties, had the highest average turnout (around 0.47) and moderate income indices. The second cluster of seven counties showed mid-range turnout (about 0.42) but the highest average income indices, indicating that the wealthiest areas did not always correspond to the absolute highest turnout. The third and largest cluster (twenty-one counties) had both the lowest turnout (approximately 0.40) and lower income indices.

Visualizing these clusters on a scatter plot of income versus turnout makes the separation immediately clear: counties in Cluster 0 sit at the top-middle of the turnout axis, Cluster 1 at the top-right on income, and Cluster 2 down toward the lower end of both scales. A complementary bar chart of mean turnout by cluster underscores that highest turnout aligns with moderate income rather than with the very wealthiest counties.

In sum, while there is a general tendency for higher‐income counties to vote more, the richest counties did not produce the very highest turnout—those honors went to a moderately affluent group. Likewise, the poorest counties exhibited the lowest participation, but income alone cannot fully explain turnout patterns. This suggests that local dynamics, demographics, or targeted mobilization efforts may play an outsized role in driving electoral engagement beyond pure economic status.

### Exploring Feature Importances in Predicting Voter Patterns by County

#### Preprocessing

First, we turn all voter lifestyle columns to binary by setting "Yes" values to 1 and null values to 0. Then, we convert all money and age columns to numerics and drop any dollar signs, commas, periods, etc. Next, we aggregate all of these columns to a county level and determine the percent of Democratic voters in 2018 in each county. Finally, we combine all columns and convert to a Pandas DataFrame. This data will then be fed into a lasso regression model so that less useful predictors will be shrunk to zero and only a handful of important features will remain.

```python
# Imports
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    when, col, regexp_replace, split,
    avg as _avg, sum as _sum
)
import pyspark.sql.functions as F

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler

# Load voter file
df_raw = spark.table("databricks_demo.final_project.wa")

# Drop columns with past voter records
vote_prefixes = ["General_", "Primary_", "PresidentialPrimary_", "AnyElection_", "OtherElection_"]
vote_feats = [c for c in df.columns if any(c.startswith(pref) for pref in vote_prefixes)]

df = df.drop(*vote_feats)

# Binarize “Yes/null” commercial flags
comm_cols = [c for c in df_raw.columns if c.startswith("CommercialDataLL_")]
df = df_raw.na.fill({c: "No" for c in comm_cols})
for c in comm_cols:
    df = df.withColumn(f"{c}_bin", when(col(c)=="Yes", 1).otherwise(0))
df = df.drop(*comm_cols)

# Convert money and age columns to numeric dtypes
## Age
df = df.withColumn("Voters_Age_num", col("Voters_Age").cast("double")).drop("Voters_Age")

## Home value
df = df.withColumn(
    "EstHomeValue_num",
    regexp_replace(col("CommercialData_EstHomeValue"), r"[\$,]", "").cast("double")
)

## Income amount
df = df.withColumn(
    "EstHHIncomeAmount_num",
    regexp_replace(col("CommercialData_EstimatedHHIncomeAmount"), r"[\$,]", "").cast("double")
)

## Convert income range to average income
df = df.withColumn(
    "IncRange", regexp_replace(col("CommercialData_EstimatedHHIncome"), r"[\$,]", "")
)
df = df.withColumn("bounds", split(col("IncRange"), "-"))
df = df.withColumn(
    "EstHHIncome_avg_num",
    (col("bounds")[0].cast("double") + col("bounds")[1].cast("double"))/2
)
df = df.drop("CommercialData_EstHomeValue",
             "CommercialData_EstimatedHHIncomeAmount",
             "CommercialData_EstimatedHHIncome",
             "IncRange","bounds")

# _bin and _num feature lists
feat_bins = [c for c in df.columns if c.endswith("_bin")]
feat_nums = [c for c in df.columns if c.endswith("_num")]

# Aggregate to county level
agg_exprs = (
    [_avg(c).alias(c + "_prop")  for c in feat_bins] +
    [_avg(c).alias(c + "_mean")  for c in feat_nums]
)
features_df = df.groupBy("County").agg(*agg_exprs)

# Build continuous dem_frac label for 2018
label_df = (
    df_raw
      .select("County",
              "ElectionReturns_G18CountyTurnoutDemocrats",
              "ElectionReturns_G18CountyTurnoutRepublicans")
      .distinct()
      .withColumn(
          "dem_pct",
          regexp_replace(
            col("ElectionReturns_G18CountyTurnoutDemocrats"), "%", ""
          ).cast("double")
      )
      .withColumn(
          "rep_pct",
          regexp_replace(
            col("ElectionReturns_G18CountyTurnoutRepublicans"), "%", ""
          ).cast("double")
      )
      .withColumn(
          "dem_frac",
          col("dem_pct")/(col("dem_pct")+col("rep_pct"))
      )
      .select("County","dem_frac")
)

# Join and drop any nulls
model_df = features_df.join(label_df, on="County", how="inner").na.drop()

# Convert dataset to Pandas
model_pd = model_df.drop("County").toPandas()
print("Model dataset has shape: ", model_pd.shape)

# Split model_pd into predictor and target dataset
X = model_pd.drop("dem_frac", axis=1)
y = model_pd["dem_frac"]
```

#### Fit Lasso Regression Model

Next, we fit a lasso regression model and extract nonzero coefficients. A lasso regression model selects a few features that are the most useful in prediction. After running the lasso regression model, we extract our non-zero coefficients to determine what voter characteristics are the most important in predicting percentage of Democratic voters by county. We standardized all predictors and used 5-fold cross-validation (via `LassoCV`) to select the L1 penalty and obtain an unbiased RMSE estimate. In this way, each county serves once as the validation fold, so we did not hold out a separate test set.

```python
# Standardize X for fitting Lasso
scaler = StandardScaler()
Xs = scaler.fit_transform(X)  # X_nv is your non-voting feature matrix

# Fit LassoCV
lasso = LassoCV(
    cv=5,
    n_alphas=100,
    eps=1e-3,
    max_iter=20000,
    tol=1e-4,
    random_state=0,
    n_jobs=-1
)
lasso.fit(Xs, y)

# Extract nonzero coefficients
coefs = pd.Series(lasso.coef_, index=X.columns)
nonzero = coefs[coefs != 0].sort_values(key=abs, ascending=False)

# Plot the top 10 features by importance
top10 = nonzero.head(10)
plt.figure(figsize=(12,6))
plt.barh(top10.index, top10.values)
plt.gca().invert_yaxis()
plt.xlabel("Lasso Coefficient")
plt.title("Top 10 Lasso Coefficients")
plt.tight_layout()
plt.show()
```

Plotted above are the nonzero coefficients of the features fed into the lasso regression model. We can see that counties with a higher average income ranges and higher home values tend to vote Democrat. We also see that older people have a tendency to vote blue. Additionally, interests like cooking, watching basketball, and woodworking are associated with Democratic voting patterns. On the other hand, interest in golf, religion, and sailing are associated with Republican voting patterns.
